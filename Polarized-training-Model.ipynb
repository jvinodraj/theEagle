The comparison between the effects of different training models — Polarized (pol), High-Intensity Interval Training (HIIT), Threshold (thr), and High-Volume Low-Intensity Training (hvt) — on various performance metrics over nine weeks for well-trained endurance athletes (including cyclists).

Here’s the difference:

1. **VO2 Max (Maximum Oxygen Uptake)**:  
   - Polarized training led to the greatest improvement at **11.7%**.  
   - HIIT training increased VO2 max by around **4%**.  
   - Threshold training (thr) actually saw a **slight decrease** in VO2 max.  
   - High-volume low-intensity (hvt) training resulted in a small positive change.  

2. **Power at 4 mmol/L Lactate (A marker for lactate threshold)**:  
   - Polarized training again showed the best improvement at **8.1%**.  
   - HIIT showed a smaller but still noticeable increase of about **4%**.  
   - Threshold and high-volume low-intensity training barely made any difference.  

3. **Power at 2 mmol/L Lactate (Another marker of aerobic efficiency)**:  
   - HIIT had the best result here, increasing power by **12.1%**.  
   - Polarized training was close behind with around **8%** improvement.  
   - Threshold and high-volume low-intensity training made only minimal changes.

```python
import pandas as pd

# Weekly mileage and distribution percentages
weekly_mileage = 50

distribution = {
    'Easy Run': 40,      # 40% of weekly mileage
    'Tempo Run': 10,     # 10% of weekly mileage
    'Interval Run': 15,  # 15% of weekly mileage
    'LSD': 45            # 45% of weekly mileage
}

# Calculate the distance for each type of run
run_distances = {run: round(weekly_mileage * (percent / 100), 1) for run, percent in distribution.items()}

# Create the detailed training plan
training_plan = [
    ['Monday', 'Easy Run', run_distances['Easy Run'] * 0.6, 'Zone 2', '8:30–9:00', '130–140 bpm'],
    ['Tuesday', 'Easy Run', run_distances['Easy Run'] * 0.3, 'Zone 2', '8:30–9:00', '130–140 bpm'],
    ['Tuesday', 'Tempo Run', run_distances['Tempo Run'], 'Zone 4', '5:00', '155–165 bpm'],
    ['Wednesday', 'Strength Training', '-', '-', '-'],
    ['Thursday', 'Interval Run', run_distances['Interval Run'], 'Zone 4-5', '4:00–4:10', '160–175 bpm'],
    ['Friday', 'Strength Training', '-', '-', '-'],
    ['Saturday', 'LSD', run_distances['LSD'], 'Zone 2', '8:30–9:00', '130–140 bpm'],
    ['Sunday', 'Rest', '-', '-', '-']
]

# Create DataFrame for training plan
plan_df = pd.DataFrame(training_plan, columns=['Day', 'Run Type', 'Distance (km)', 'Zone', 'Pace (min/km)', 'HR (bpm)'])

# Create summary table
summary_df = pd.DataFrame(
    [[run, percent, km] for run, (percent, km) in zip(distribution.keys(), zip(distribution.values(), run_distances.values()))],
    columns=['Run Type', 'Percentage (%)', 'Distance (km)']
)

# Display the tables
print("Training Plan:\n")
display(plan_df)

print("\nSummary Table:\n")
display(summary_df)
```
